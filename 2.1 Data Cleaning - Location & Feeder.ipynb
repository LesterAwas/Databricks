# Data Cleaning - Location & Feeder


### Locations Silver
Separate Location and Feeder data from initial bronze table

```python
%sql
CREATE OR REPLACE TABLE beneco_pipeline.silver.location_silver
AS SELECT 
  Barangay,
  Municipality
FROM beneco_pipeline.bronze.loc_feeder_bronze;
```

### Feeder Silver
The next cell gets all feeder names from the loc_feeder_bronze table.

Each row in the table shows a barangay and the feeder or feeders that supply it. Sometimes, a barangay has more than one feeder, in this case the feeder names are shown as a comma-separated string.

To extract the feeder names, 
* The feeder names are first split using `, `.
* Each feeder name is then placed on its own row.
* Lastly, duplicate names are removed by selecting only distinct feeder names.

```python
%sql
CREATE OR REPLACE TABLE beneco_pipeline.silver.feeder_silver
AS SELECT DISTINCT
  Feeder
FROM (
  SELECT 
    explode(split(Feeder, ', ')) AS Feeder
  FROM beneco_pipeline.bronze.loc_feeder_bronze)
ORDER BY Feeder;
```

```python
%sql
SELECT * FROM beneco_pipeline.silver.feeder_silver;
```
