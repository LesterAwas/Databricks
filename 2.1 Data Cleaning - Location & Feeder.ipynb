```python
%sql
CREATE OR REPLACE TABLE beneco_pipeline.silver.location_silver
AS SELECT 
  Barangay,
  Municipality
FROM beneco_pipeline.bronze.location_bronze;

SELECT DISTINCT * FROM beneco_pipeline.silver.location_silver;
```

```python
%sql
CREATE OR REPLACE TABLE beneco_pipeline.silver.feeder_silver
AS SELECT DISTINCT
  Feeder
FROM (
  SELECT 
    explode(split(trim(Feeder), ', ')) AS Feeder
  FROM beneco_pipeline.bronze.location_bronze
);
SELECT * FROM beneco_pipeline.silver.feeder_silver ORDER BY Feeder;
```
