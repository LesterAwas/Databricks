# Ingestion Pipeline

### Interruption Data
Find all csv files in `/Volumes/beneco_pipeline/bronze/data/Interruption` and save into a list

*Note Data was scraped from posts on BENECO facbook page using Appify

```python
import os

directory_path = "/Volumes/beneco_pipeline/bronze/data/Interruption"
csv_files = [f"{directory_path}/{f}" for f in os.listdir(directory_path) if f.lower().endswith('.csv')]

# Shows all detected CSV files in the folder
print(csv_files)  
```

Batch read all csv files in list, remove duplicates and add to `interruption_bronze` table

```python
df = (
    spark.read
    .format("csv")
    .option("header", "true")
    .option("multiline", "true")
    .option("escape", '"')
    .option("quote", '"')
    .load(csv_files)
)
df = df.dropDuplicates()
df.write.mode("overwrite").saveAsTable("beneco_pipeline.bronze.interruption_bronze")
```

### Locations and Feeder Data
Location and feeder data taken directly from BENECO know your feeder site https://www.beneco.com.ph/feeders.php

```python
df = (
    spark.read
    .format("csv")
    .option("header", "true")
    .option("multiline", "true")
    .option("escape", '"')
    .option("quote", '"')
    .load("/Volumes/beneco_pipeline/bronze/data/loc_feeder.csv")
)
df.write.mode("overwrite").saveAsTable("beneco_pipeline.bronze.location_bronze")
```
